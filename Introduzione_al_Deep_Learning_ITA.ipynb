Definizioni

Una rete neurale (**neural network** - NN)  un tipo di un sistema computazionale ispirato alla struttura e al funzionamento del cervello umano.  composta da migliaia o addirittura milioni di semplici nodi di elaborazione, chiamati neuroni ( "neurons") densamente interconnessi, che imitano vagamente i neuroni biologici del nostro cervello.

Questi neuroni artificiali sono disposti in strati (*layer*), generalmente comprensivi di:

- un livello di ingresso, detto Input Layer;

- uno o pi livelli profondi che eseguono i calcoli principali della rete, Hidden Layers;  

- un livello di uscita, Output Layer.

Ogni neurone artificiale  collegato ad altri neuroni. Ad ogni connessione  associato un *weight*. Questi regolano il flusso di informazioni nella rete, in pratica determinando l'importanza di ogni connessione.  Inoltre, ogni neurone ha una funzione di attivazione; che definisce un criterio per controllare l'ampiezza dell'output di un neurone.

Attraverso un processo chiamato "training", in cui vengono regolati i *weights*, la rete neurale impara a svolgere compiti specifici.

Matematicamente un certo neurone $k$ (detto anche **neuron** with label $k$) pu essere cos descritto:

$u_{k} = w_{k1}x_{1} + w_{k2}x_{2} + w_{k3}x_{3} + ... + w_{kn}x_{m} + b_{k}= \sum_{i=1} ^{m} w_{ki}x_{i} +  b_{k}$


dove:

$(x_{1},x_{2},x_{3},...,x_{m})$ sono le varaibili per indicare i segnali di input.

$(w_{k1},w_{k2},...,w_{km})$ sono le variabili per indicare i pesi relativi al neuorone $k$

$b_{k}$  detto bias;  una costante usata come offset (in pratica ha l'effetto applicare una traslazione al risultato della somma $\sum_{i=1} ^{m} w_{ki}x_{i}$. Per illustrare meglio il concetto, si prenda in considerazione il caso $m=1$, e la relativa retta $wx + b$)

Si noti:

$\sum_{i=1} ^{m} w_{ki}x_{i} +  b_{k}$  una combinazione lineare dei segnali di input pi il bias.


$\;$

Sia $ \phi (\cdot) $ la funzione di attivazione, allora il segnale di output $y_{k}$ del neurone $k$,  cos descritto:


 $y_{k} = \phi ( \sum_{i=1} ^{m} w_{ki}x_{i} +  b_{k} )$

Si noti che, in una rete neurale multistrato l'uscita di un certo neurone pu essere  uno dei segnali di input di un altro neurone.


Perche una funzione di attivazione?

Per modellare uno scenario non lineare (non descrivibile matematicamente come una combinazione lineare), dobbiamo introdurre direttamente una " non linearit", ovvero $ \phi (\cdot) $ , la funzione di attivazione

Ci sono diverse tipologie di funzione di attivazione, le pi comuni sono: threshold function, sigmoid function, rectified linear unit (ReLu).

Un esenpio sigmoid function  la logistic
function:

$ \phi ( \sum_{i=1} ^{m} w_{ki}x_{i} +  b_{k} ) =  \frac{1}{1+e^{ - a ( \sum_{i=1} ^{m} w_{ki}x_{i} +  b_{k} ) }}$

dove $a$  una costante detta slope parameter, variando i valori di tale parametro otteniamo funzioni sigmoidali di diverse pendenze.

${\displaystyle x}$: il vettore di input

In ternimi generali, un intera rete neurale pu essere descritta matemanticamente coma una combinazione di composizione di funzioni e moltiplicazione di matrici:


${\displaystyle y(x):=f^{L}(W^{L}f^{L-1}(W^{L-1}\cdots f^{1}(W^{1}x)\cdots ))}$

**Universal approximation theorems**: ovvero le reti neurali possono rappresentare qualsiasi funzione?

Le capacit di approssimazione delle reti neurali multistrato sono state oggetto di numerosi studi scientifici.

Nella teoria matematica delle reti neurali artificiali, i teoremi di approssimazione universale sono risultati che pongono limiti a ci che le reti neurali possono teoricamente apprendere e rappresentare.



Il teorema di approssimazione universale (George Cybenko e Kurt Hornik ) afferma che una rete neurale con un solo hidden layer contenente un numero finito di neuroni (con relative non-linear activation function) pu approssimare con un qualsiasi grado di precisione una qualsiasi funzione continua su un compatto.

In parole pi semplici, questo significa che le reti neurali multistrato sono in grado di modellare un'ampia variet di fenomeni, purch siano descritti da funzioni continue.


Ref:

https://en.wikipedia.org/wiki/Universal_approximation_theorem

**Learning process**


Il processo di apprendimento (learning process) delle  reti neurali mira a modellare il comportamento della rete su un determinato compito, aggiornando opportunamente il vaolre dei pesi.



Il processo di apprendimento pi comune  un ciclo iterativo che pu essere suddiviso in tre parti principali:

- Forward propagation
- Calcolo della funzione costo (o detta anche loss function)
- Backward propagation


$\;$

Fase preliminare:

Creazione di un training dataset ad hoc.


$\;$


Forward propagation:

Definita l'architettura della rete neurale, inseriano in questa i dati di input del training dataset; da qui inizia un processo di elaborazione che coinvolge i tutti i nodi della rete; dando in uscita uno o pi valori numerici, detti predizione.


$\;$

Calcolo della funzione costo (loss function):

La loss function, serve a misurare la discrepanza tra i valori in uscita dalla rete, ovvero le predizioni, e i valori attesi (di fatto misura l'errore di predizione commesso dalla rete). L'obiettivo principale di una loss function  quello di guidare il processo di apprendimento della rete neurale.

Esistono diverse tipologie di loss function, ognuna con caratteristiche e scopi specifici. Tra le pi utilizzate c' la Mean Squared Error (MSE): che misura la distanza quadratica media tra le predizioni e i valori attesi.


$L_{\text{MSE}}(y, \hat{y})  =  \frac{1}{n} ( (y_{1} -  \hat{y}_{1})^2 + (y_{2} -  \hat{y}_{2})^2 + ... + (y_{n} -  \hat{y}_{n})^2 )   $

Dove

$y_{i}$  il valore atteso fornito dal dataset

$\hat{y}_{i}$ rappresenta la predizione in uscita della NN. Si ricorda che tale valore  in funzione dei pesi della rete, e dei segnali di input (associati la valore atteso $y_{i}$) inseriti.

$n$  il numero dei training example (input+valore_atteso) elaborati. Nel caso del codice di esempio sottostante $n=4$.

Per migliorare le prestazini della rete neurale dobbiamo minimizzare il valore della funzione costo ottenuto alla fine di ogni iterazione, aggiornando opportunamente i pesi. Questo si concretizza durante la fase di backpropagation.

$\;$


Backward propagation


Tale fase consente alla rete di aggiornare i propri pesi  in modo da ridurre l'errore di predizione in uscita (rispetto i valori attesi).

La modifica dei pesi si pu ottenere con tecniche di ottimizzazione matematica.
La tecnica che tipicamente viene utilizzata  quella della discesa del gradiente o Gradient Descent (usata nel codice di esempio sottostante). Di base, si sfrutta l'informazionie ottenuta dal gradiente della funzione loss per trovare il minimo di tale funzione, o avvicinarsi il pi possibile ad esso in tempi accettabili.



Nella pratica, per una coppia specifica di input e valore atteso, calcoliamo il gradiente della funzione di loss rispetto ai singoli pesi. Per fare questo, ci avvaliamo della formula della "chain rule" delle derivate. Poi, per ridurre l'errore di predizione, aggiorniamo il peso corrente sottraendo il valore ottenuto dal gradiente, in genere moltiplicato per un certo learning rate, al peso stesso:



$w_{j}  w_{j} -  \frac{\partial{L (y, \hat{y} (\bar{w} ) ) } }{\partial{w_{j}} } $

dove:

$ > 0 $  il learning rate.

$\hat{y}(\bar{w})$ sottolinea il fatto che l'output predetto dalla rete sia in funzione dei pesi $\bar{w}$


Cos facendo, l'output della rete neurale si avvicina al valore atteso, con leffetto di ridurre il valore della loss function ad ogni iterazione.

In genere, si ripete il processo iterativo nel suo dataset finch la loss fuction non smette di cambiare o almeno cambia in modo molto lento. In questo caso, diciamo che il modello  convergente.

##Implementare una rete neurale passo per passo


Vediamo come una rete neurale impara a svolgere l'operazione della porta logica XOR.
Si ricorda che lo XOR restituisce 1 solo quando esattamente uno dei suoi ingressi  1, altrimenti restituisce 0.

Per eseguire l'operazione di  XOR, costruiamo una semplice rete neurale:
 - due nodi di ingresso,
 - un hidden layers di cinque nodi,
 - un layer di output composto da un nodo

```python
# codice ispirato da : https://github.com/PacktPublishing/Deep-Reinforcement-Learning-with-Python/blob/master/07.%20Deep%20learning%20foundations/7.05%20Building%20Neural%20Network%20from%20scratch.ipynb

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Premessa tecnica per capire il codice principale


```python
# Dot product

h1 = [[1, 2], [3, 4]]
h2 = [[5, 0], [6, 7]]

r = np.dot(h1, h2) # cio :
# r11 1*5 + 2*6 = 17
# r12 = 1*0 + 2*7 = 14
# r21 = 3*5 + 4*6 = 39
# r22 = 3*0 + 4*7 = 28
print(r)
```

```python
# computation of forward propagation

def sigmoid(z):
    return 1 / (1+np.exp(-z))

def sigmoid_derivative(z):
    return np.exp(-z)/((1+np.exp(-z))**2)

num_input = 2
num_hidden = 5
num_output = 1

# Dataset di input
x = np.array([ [0, 1], [1, 0], [1, 1],[0, 0] ]) #  # 4 elementi a due input [x1,x2]

Wxh = np.ones((num_input,num_hidden))*3
print("Wxh =\n",Wxh)
Why = np.ones((num_hidden,num_output))*2
print("Why =\n",Why)
bh = np.zeros((1,num_hidden))
print("bh =",bh)
by = np.zeros((1,num_output))

z = np.dot(x,Wxh) + bh
#Formula di base
# z = input x1 * valore del peso wk1 del neurone k + input x2 * valore del peso wk2 del neurone k + valore del bias
#
# input [0, 1] , neuron with label k = 1
# z11 = x1*w11 + x2*w21 = 0*3 + 1*3 = 3
# input [0, 1] ,neuron with label k = 2
# z12 = x1*w12 + x2*w22 = 0*3 + 1*3 = 3
# input [0, 1] , neuron with label k = 3
# ...
# input [1, 0] , neuron with label k = 1
# z21 = 1*3 + 0*3 = 3
# ...
# input [1, 1] , neuron with label k = 1
# z31 = 1*3 + 1*3 = 3
# ...
#

# La matrice z contiene tutti i valori delle uscite dei 5 neuroni (senza activation function)
# ottenuti dando in ingresso l'intero dataset di input.
print("Valori dei 5 neuroni della hidden layer (senza act funct).")
print("Ogni riga si riferisce ad uno dei 4 valori di input [x1,x2] immessi, ogni colonna si riferisce ad uno dei 5 neuroni usati \n", z)

a = sigmoid(z)
print("Usando l'activation function, output della hidden layer diventa = \n", a)

Output_layer = np.dot(a,Why) + by
# out_layer = risultato di uscita del neurone k * peso della output layer + bias
# questo per tutti e 5 i neuroni

y_out = sigmoid(Output_layer)
print("Valori in uscita della NN.")
print("Ogni riga si riferisce ad uno dei 4 valori di input [x1,x2] immessi:\n", y_out )
```

```python
# cost function

y_target = np.array([ [1], [1], [0], [0]])
print("y_target \n",y_target)
print("y_out\n ",y_out)
print("y_target-y_out)**2 \n" , (y_target-y_out)**2)

print("Sum(y_target-y_out)**2) = " , sum((y_target-y_out)**2))

J = 1/len((y_target-y_out))*sum((y_target-y_out)**2) # nota : len((y_target-y_out) = 4
print("MSE = ", J)
```

```python
# computation of backword propagation

# La formula della "chain rule"  il metodo qui usato per calcolare la derivata di una funzione composta
# In questo caso, siano  "w_h_y" i 5 pesi che collegano hidden layer con output layer.
# Per calcolare la derivata della Loss J, rispetto tali pesi, sfruttiamo la formula:
# dJ / dw_h_y  =  dJ / d y_out  *  d y_out / d_hidden_layer_output  *  d_hidden_layer_output / w_h_y
#
print("Derivata della MSE rispetto y_out \n", -0.5*(y_target - y_out))
print("Sigmoid_derivative(Output_layer)\n", sigmoid_derivative(Output_layer))
delta2 = np.multiply(-0.5*(y_target - y_out),sigmoid_derivative(Output_layer))
print("delta2 = np.multiply(Derivata della MSE rispetto y_out, sigmoid_derivative(Output_layer))\n", delta2)

print("Sia  d_hidden_layer_output / w_h_y  = Trasposta dell'output dei 5 nodi della hidden layer\n",a.T)
dJ_dWhy = np.dot(a.T, delta2)
print("d_J / d_w_h_y = np.dot(d_hidden_layer_output / w_h_y , delta2)\n", dJ_dWhy)

print("")

# Analoga procedura per i 10 pesi tra input ed hidden layer
delta1 = np.dot(delta2,Why.T)*sigmoid_derivative(z)
dJ_dWxh = np.dot(x.T, delta1)
print("dJ / dw_x_h =\n", dJ_dWxh)
```

### Fase di Training

```python
# Definiamo la funzione di attivazione sigmoide e la rispettiva derivata
def sigmoid(z):
    return 1 / (1+np.exp(-z))

def sigmoid_derivative(z):
    return np.exp(-z)/((1+np.exp(-z))**2)


#  forward propagation :
def forward_prop(x,Wxh,Why):
    z1 = np.dot(x,Wxh) + bh
    a1 = sigmoid(z1)
    z2 = np.dot(a1,Why) + by
    y_out = sigmoid(z2)
    return z1,a1,z2,y_out

#  backward propagation:
def backword_prop(y_out, z1, a1, z2):
    delta2 = np.multiply(-0.5*(y_target - y_out),sigmoid_derivative(z2)) # Nota, se pur non prorpio teoricamente ortodosso,
    # sostituendo il parametro -0.5 , ottenuto dall'operazione di derivazione 2 * 0.25 , con il valore -3
    # l'algoritmo converge pi velocemente.
    dJ_dWhy = np.dot(a1.T, delta2)
    delta1 = np.dot(delta2,Why.T)*sigmoid_derivative(z1)
    dJ_dWxh = np.dot(x.T, delta1)
    return dJ_dWxh, dJ_dWhy

# cost function MSE
def cost_function(y_target, y_out):
    J = 1/len((y_target-y_out))*sum((y_target-y_out)**2)
    return J


# Note :
# Batch gradient descent
# I parametri del modello vengono aggiornati solo dopo che tutti gli esempi
# del dataset training sono stati elaborati.
 # In pratica : ogni esempio del training dataset contribuisce alla stima J per ogni episodio.


# Per simulare le operazione di XOR, abbiamo bisogno di un training Dataset, ovvero di:
# dati di ingresso
x = np.array([ [0, 1], [1, 0], [1, 1],[0, 0] ])  # 4 elementi a due input [x1,x2]
# e relativi dati di uscita attesi
y_target = np.array([ [1], [1], [0], [0]])

# Definiamo il numero di nodi in ciascuno strato:
num_input = 2
num_hidden = 5
num_output = 1

# Inizializza pesi e bias in modo casuale.
print("Valore dei pesi in entrata, inizio training")
Wxh = np.random.randn(num_input,num_hidden)
print("Wxh =",repr(Wxh))
Why = np.random.randn (num_hidden,num_output)
print("Why =",repr(Why))
bh = np.zeros((1,num_hidden))
by = np.zeros((1,num_output))


alpha = 0.4   # learning rate, determina la velocit con cui la rete aggiorna i suoi pesi durante il processo di training.
# Un valore troppo alto o troppo basso pu rallentare l'apprendimento o addirittura impedirgli di convergere.

num_iterations = 10_000

# trainig loop

cost = []
for i in range(num_iterations):

    #perform forward propagation and predict output
    z1,a1,z2,y_out = forward_prop(x,Wxh,Why)

    #perform backward propagation and calculate gradients
    dJ_dWxh, dJ_dWhy = backword_prop(y_out, z1, a1, z2)

    #update the weights
    Wxh = Wxh -alpha * dJ_dWxh
    Why = Why -alpha * dJ_dWhy

    #compute cost
    c = cost_function(y_target, y_out)

    #store the cost for plot
    cost.append(c)

print("\nFine del training")
print("Wxh =",repr(Wxh))
print("Why =",repr(Why))
```

```python
 # Grafico della  cost function:
plt.figure(figsize=[6, 3])

plt.grid()
plt.plot(range(num_iterations),cost)

plt.title('Cost Function')
plt.xlabel('Training Iterations')
plt.ylabel('Cost')
```

```python
# Test della rete ottenuta

test = [0,0]
_,_,_, out = forward_prop(test,Wxh,Why)
print(test,"->",out.round(2))


test = [1,0]
_,_,_, out = forward_prop(test,Wxh,Why)
print(test,"->",out.round(2))


test = [0,1]
_,_,_, out = forward_prop(test,Wxh,Why)
print(test,"->",out.round(2))

test = [1,1]
_,_,_, out = forward_prop(test,Wxh,Why)
print(test,"->",out.round(2))

```

L'output della rete neurale ottenuta riproduce piuttosto bene il funzionamento di una porta XOR.
