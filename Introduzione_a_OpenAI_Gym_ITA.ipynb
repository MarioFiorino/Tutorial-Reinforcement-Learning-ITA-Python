# Descrizione e comprenzione dell'ambiente Frozen Lake

Come mostrato in figura sopra, Frozen Lake, è una griglia di 16 stati, con le seguenti caratteristiche:

* uno stato iniziale (lo stato start: nell'angolo alto a sinistra)

- una serie di stati semplici(ghiacciati)

- una serie di stati con buco(buchi nel ghiaccio)

- uno stato obbiettivo che da una ricompensa (il regalo)


L'agente puo muoversi attraverso 4 possibili azioni, che sono su, giù, sinistra e destra.

Gli spostamenti da uno stato all'altro sono di natura probabilistica. Questo significa che lo stato successivo non è perfettamente prevedibile data la conoscenza dello stato precedente e dell'azione dell'agente. In pratica l'agente puo "scivolare" , ed anzichè andare nella direzione dell'azione selezionata "destra", può andare a "sù" oppure "giù". Ha il 33% di possibilità di andare effetivamente nella direzione selezionata.

L'agente deve iniziare dallo stato Start e raggiungere lo stato obbiettivo. Ma ha un problema, se l'agente visita lo  stato del buco, allora l'agente cadrà nel buco e morirà.

Se raggiunge l'obbiettivo, l'agente avra un reward di +1. Per il resto degli stati il reward è 0.

Nel toolkit di Gym :

I nostri stati (da Start a Goal) saranno codificati da 0 a 15

Tutte e quattro le azioni: sinistra, giù, su, destra saranno codificate da 0 a 3, esattamente :

0 -> sinistra

1 -> giù

2 -> destra

3 -> sopra

```python
#La libreria Gym ci consente di eseguire alcune query per avere una sintesi dell'ambiente

def query_environment(name):
    env = gym.make(name)
    spec = gym.spec(name)
    print(f"Action Space: {env.action_space}")
    print(f"Observation Space: {env.observation_space}")
    print(f"Max Episode Steps: {spec.max_episode_steps}")
    print(f"Reward Range: {env.reward_range}")
```

```python
query_environment('FrozenLake-v1')
```

### Definizione dell'ambiente

Frozen Lake è un ambinete MDP (Markov decision process), caratterizzato da:

- Stati: un insieme di stati presenti nell'ambiente
- Azioni: una serie di azioni che l'agente può eseguire in ogni stato.
- Probabilità di transizione - La probabilità di transizione è indicata con $P(s'|s,a)$, ed indica la probabilità di spostarsi da uno stato $s$ allo stato $s'$ mentre si esegue un'azione $a$

- Funzione di ricompensa: è la ricompensa che l'agente ottiene spostandosi in certo uno stato


```python
# env.observation_space
# La struttura di base dell'ambiente è descritta da env.observation_space (ci da in output le caratteristiche dello spazio degli stati del ambiente)
# Nonchè può generare un campionamento di uno stato.

print(env.observation_space)
nS = env.observation_space.n # Dimensioni dello spazio delgi stati
print("Numero di stati totali nel nostro ambiente : ", nS)
print("*\n")

print(dir(env.observation_space))
print("*\n")

print("Stato campionato random: ", env.observation_space.sample())
```

Il nostro ambiente Frozen Lake è stocastico. Questo significa che, a seconda di fattori casuali, le azioni che eseguiamo possono avere risultati diversi.

Ad esempio, se siamo nello stato 2 e eseguiamo l'azione 1 (giù), possiamo raggiungere lo stato 6. Tuttavia, non è sempre così. Con una probabilità di 0,33333, possiamo raggiungere lo stato 1; e con la stessa probabilità, possiamo raggiungere lo stato 3.

```python
#env.P[state][action]
#In un ambiente stocastico, le azioni che eseguiamo possono portare a stati diversi con una certa probabilità.
#Per ottenere informazioni sulla probabilità di transizione e sul reward,
# possiamo sfruttare l'istruzione: env.P[state][action]

# Format di output: [(transition probability, next state, reward, Is terminal state?)]
print("La lista di tuple env.P[state = 2][action = 1] si presenta così:\n",env.P[2][1])
print(type(env.P[2][1]))
print("nr elementi in lista : ", len(env.P[2][1]))

print("\n***\n")
print("Un modo più leggibile di gestire env.P[state][action] \n ")

format = ("transition probability", "next state", "reward", "Is terminal state?")
for i in range(0,(len(env.P[2][1]))):
  for j in range(0,4):
    print(format[j]," : ",env.P[2][1][i][j])
  print(" ")
```

## Generare un episodio in Frozen Lake



```python
observation = env.reset()

plt.rcParams["figure.figsize"] = [6.00, 3.00] # per settare l'immagine

for t in range(30):
    plt.imshow(env.render(mode='rgb_array'),extent=[0, 4, 4, 0])
    display.display(plt.gcf())

    #Se si vuole provare ad interagire in prima persona
    #action = input("\n 0 : < sinistra || 1 : v giù || 2 : > destra  || 3 : ^ sopra\n  ")
    #action = int(action)

    # Per un campionamento random dell'azione
    #action = env.action_space.sample()

    #La migliore policy da eseguire per provare ad arrivare all'obbiettivo
    #Per policy si intende la funzione: State -> Action
    policy = [0, 3, 3, 3, 0, 0, 0, 0, 3, 1, 0, 0, 0, 2, 1, 0]
    #ovvero un forma grafica
    #sfruttando il dict a2w = {0:'<', 1:'v', 2:'>', 3:'^'}
    #['<' '^' '^' '^']
    #['<' '<' '<' '<']
    #['^' 'v' '<' '<']
    #['<' '>' 'v' '<']
    # Nota gli stati terminali(buche+goal) hanno tutti valore 0

    action = policy[observation]

    observation, reward, done, info = env.step(action)
    print(observation, reward, done)
    #display.clear_output(wait=True)
    if done: # Se il gioco viene terminato
        print("Episodio terminato dopo {} passaggi temporali".format(t+1)) # Printa i timesteps
        plt.imshow(env.render(mode='rgb_array'),extent=[0, 4, 4, 0])
        break
```
